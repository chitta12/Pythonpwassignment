# Question 2 : Explain the difference between List and Tuple in Python.

1. Mutability (The Core Difference)
  Lists are Mutable: Once you create a list, you can alter its contents. You can add new items, remove existing ones, or change the value of specific
  elements.

 Tuples are Immutable: Once you create a tuple, it is locked. You cannot add, remove, or modify its elements. If you need to change a tuple, you    
 essentially have to create a brand new one.

2. Syntax
  Lists are defined using square brackets: my_list = [1,2,3]

  Tuples are defined using parentheses: my_tuple = (1, 2, 3)

3. Performance and Memory
  Because lists need to allocate extra memory to accommodate potential future additions or changes, they take up slightly more space and are slightly
  slower.

  Because tuples are fixed in size, Python can optimize them better. They consume less memory and are slightly faster to iterate through.

4. Available Methods
  Lists come with a wide variety of built-in methods to modify them, such as .append(), .remove(), .pop(), and .sort().

  Tuples have a very restricted set of methods since they cannot be modified. They only support .count() (to count occurrences of an item) and .index()   (to find the position of an item).

# Question 7 : Create a list of five numbers and print the maximum number from the list. 

```python
while True:
    numbers = input("Enter exactly 5 numbers separated by commas: ").split(",")

    if len(numbers) != 5:
        print("Please enter exactly 5 numbers!")
    else:
        numbers = list(map(int, numbers))
        break

print(f"The list is: {numbers}")
print(f"Maximum number is: {max(numbers)}")
```

# Question 8 : Write a Python program to remove duplicate values from a list using a set. 

```python
numbers = [1, 2, 2, 3, 4, 4, 5]

unique_numbers = list(set(numbers))

print(f"Original list: {numbers}")
print(f"List after removing duplicates: {unique_numbers}")
```

# Question 9 : Write a function that returns the square of a number. 

```python
def square(num):
    return num * num

number = float(input("Enter a number: "))

result = square(number)

print(f"Square of {number} is: {result}")
```

# Question 10 : Write a Python program to count how many times a number appears in a list.

## Example List    [2,3,4,2,5,2]

```python
numbers = [2, 3, 4, 2, 5, 2]

num = int(input("Enter the number to count: "))

count = numbers.count(num)

print(f"{num} appears {count} times in the list")
```
