Q1.Create a Pandas DataFrame using a Python dictionary with the following columns:Product,Region,Sales,Quantity

```python
import pandas as pd

#  Create a dictionary with the data
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Region': ['North', 'South', 'East', 'West'],
    'Sales': [1200.00, 25.50, 45.00, 300.00],
    'Quantity': [10, 150, 85, 20]
}

#  Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(data)

#  Display the result
print(df)
```

Q4. Indexing using .loc : Using .loc, perform the following:

Display the row with index 10.

Display the Sales value of the row with index 20


```python
# Displaying the row with index 10 Using .loc.
row_with_index_10 = df.loc[10]
print(row_with_index_10)

print("*"*60)

#Displaying the Sales value of the row with index 20  Using .loc
sales_value_row_20 = df.loc[20, 'Sales']
print(f"Sales value of the row with index 20 is : {sales_value_row_20}")
```

Q5. Indexing using .iloc

Q7. Filtering Data

Filter the dataset to display:

Q9. Descriptive Statistics

Generate summary statistics using:

describe()

From the result identify:

Mean sales value

Maximum quantity sold

Minimum sales value

```python
# Generate summary statistics using: describe() and printing Mean sales value,Maximum quantity sold,Minimum sales value
summary_stats = df.describe()
print(summary_stats)
print("*" * 60)

# printing Maximum quantity sold
max_quantity = df['Quantity'].max()
print(f"Maximum quantity sold is : {max_quantity}")
print("*" * 60)

# printing Mean sales value
mean_sales = df['Sales'].mean()
print(f"Mean sales value is : {mean_sales}")
print("*" * 60)

# printing Minimum sales value
min_sales = df['Sales'].min()
print(f"Minimum sales value is : {min_sales}")


```

Q10. Aggregation Analysis

Using the Sales column, calculate:

Total sales

Average sales

Maximum sales

Minimum sales

```python
# calculting Total sales using Sales column
total_sales = df['Sales'].sum()
print(f"Total sales is : {total_sales}")

# calculting Average sales using Sales column
average_sales = df['Sales'].mean()
print(f"Average sales is : {average_sales}")

# calculting Maximum sales using Sales column
max_sales = df['Sales'].max()
print(f"Maximum sales is : {max_sales}")

# calculting Minimum sales using Sales column
min_sales = df['Sales'].min()
print(f"Minimum sales is : {min_sales}")
```
