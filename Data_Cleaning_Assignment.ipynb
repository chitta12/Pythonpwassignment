Q2. Handling Missing Age

Scenario:
 Age is critical for medical analysis, but some values are missing.

Task:

Replace missing Age values with an appropriate method

Justify your choice (mean/median)


```python
#Replace missing Age values with an median
age_median = df['Age'].median()
print(f"Calculated Median Age: {age_median}")
df['Age'].fillna(age_median, inplace=True)
print(df['Age'])
```

**Justification**

Median is the better choice here because age is typically represented as a whole number. Imputing with the median (50) keeps the data clean and consistent as an integer, whereas the mean would introduce a fractional age (49.59) which isn't standard in patient records.

Scenario:
 Treatment cost is highly skewed due to expensive treatments.

Task:

Handle missing Treatment_Cost values

Choose the correct imputation method and explain why

```python
cost_median = df['Treatment_Cost'].median()
print(f"Calculated Median Treatment Cost: {cost_median}")
df['Treatment_Cost'].fillna(cost_median, inplace=True)
print(df['Treatment_Cost'])
```

The median represents the exact middle point of the data when sorted from lowest to highest. It is completely unaffected by extreme outliers.

Q6. Outlier Detection (Treatment Cost)

Scenario:
 Extreme treatment costs are affecting analysis.

Task:

Detect outliers using IQR method

Display number of outliers

```python


# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Treatment_Cost'].quantile(0.25)
Q3 = df['Treatment_Cost'].quantile(0.75)

# 3. Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

#  Define the lower and upper bounds for normal values
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#  Filter the dataset to find rows outside these bounds
outliers = df[(df['Treatment_Cost'] < lower_bound) | (df['Treatment_Cost'] > upper_bound)]

#  Display the total number of outliers
print(f"Number of Outliers in Treatment_Cost: {len(outliers)}")
print(f"Any cost above ${upper_bound:,.2f} is considered an outlier.")
```

Q7. Outlier Treatment

Scenario:
The business team wants to retain all records.

Task:

Apply capping (Winsorization) on Treatment_Cost

Use 5th and 95th percentile

```python
# (Assuming missing values are already filled with the median as we did in Q3)
cost_median = df['Treatment_Cost'].median()
df['Treatment_Cost'].fillna(cost_median, inplace=True)

# 2. Calculate the 5th and 95th percentiles
p05 = df['Treatment_Cost'].quantile(0.05)
p95 = df['Treatment_Cost'].quantile(0.95)

print(f"Lower Cap (5th Percentile): {p05}")
print(f"Upper Cap (95th Percentile): {p95}")

# Apply capping using the .clip() method
# We overwrite the original column with the capped values
df['Treatment_Cost'] = df['Treatment_Cost'].clip(lower=p05, upper=p95)

# 4. Verify the new Minimum and Maximum
print(f"New Minimum: {df['Treatment_Cost'].min()}")
print(f"New Maximum: {df['Treatment_Cost'].max()}")
```

Q8. Transformation

Scenario:
 Treatment cost is highly skewed.

Task:

Apply log transformation

Create a new column

Compare before vs after distribution

```python


# Fill missing values (NEW WAY - No Warning)
cost_median = df['Treatment_Cost'].median()
df['Treatment_Cost'] = df['Treatment_Cost'].fillna(cost_median)

# Apply Log Transformation
df['Treatment_Cost_Log'] = np.log1p(df['Treatment_Cost'])

# Compare the Skewness
skew_original = df['Treatment_Cost'].skew()
skew_transformed = df['Treatment_Cost_Log'].skew()

print(f"Original Skewness: {skew_original:.2f}")
print(f"Log Transformed Skewness: {skew_transformed:.2f}")
```

```python
#  Convert 'Admission_Date' from text to actual Datetime objects
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])

#  Sort the dataset chronologically (oldest to newest)
df = df.sort_values(by='Admission_Date')

#  Apply Forward Fill (ffill)
# We use ffill() to propagate the last valid observation forward
df_time_cleaned = df.fillna(method='ffill')

# Optional: If the very first row happens to be a missing value, ffill won't catch it
# (because there is no "past" to pull from). We can chain a bfill() just for that edge case.
df_time_cleaned = df_time_cleaned.fillna(method='bfill')

#  Verify the results
print("Missing values remaining:")
print(df_time_cleaned.isnull().sum())
```

**Justification**

we have to use Forward Fill (ffill).Forward filling safely uses past data to fill gaps, mimicking reality and avoiding data leakage. In contrast, backward filling pulls future data backward, causing data leakage that severely flaws predictive models.
