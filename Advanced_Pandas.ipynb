Q2 Sales Filtering Analysis

Filter the dataset to display:

Orders where Sales > 3500

Orders where Region = "East" and Sales > 2500

Explain what insights you observe

```python
print("Orders where Sales > 3500:")
sales_above_3500 = df[df["Sales"] > 3500]
print(sales_above_3500)
print("*"*100)


print("\nOrders where Region = 'East' and Sales > 2500:")
east_sales_above_2500 = df[(df["Region"] == "East") & (df["Sales"] > 2500)]
print(east_sales_above_2500)
print("*"*100)

# observing insights
print(f"Total Orders: {len(sales_above_3500)}")
print("*"*100)
print (f"Top Category: {df['Category'].value_counts()}")
print("*"*100)
print(f"City:{df['City'].value_counts()}")
print("*"*100)
print (f"Total sales: {df['Sales'].sum()}")
```

Q3 Customer Purchase Analysis

Identify the top 5 customers based on total sales.

Hint: Use GroupBy with aggregation.

```python
#Identify the top 5 customers based on total sales.
top_customers = df.groupby("Customer")["Sales"].sum().nlargest(5)
print(top_customers)
```

Q4 Regional Sales Performance

Calculate the total sales generated by each region.

Then identify:

Which region generated the highest revenue?

```python
#total sales generated by each region
total_sales_by_region = df.groupby("Region")["Sales"].sum()
print(total_sales_by_region)
# highest revenue genrated region
highest_revenue_region = total_sales_by_region.idxmax()
print(f"\nRegion with the highest revenue: {highest_revenue_region}")
print(f"\nRegion with the highest revenue amount: {total_sales_by_region.max()}")
```

Q5 Product Performance Analysis

Calculate the total quantity sold for each product.

Which product appears to be the most popular?

```python
#total quantity sold for each product
total_quantity_by_product = df.groupby("Product")["Quantity"].sum()
print(total_quantity_by_product)

# most popular Product
most_popular_product = total_quantity_by_product.idxmax()
print(f"\nMost popular product: {most_popular_product}")
```

Q7 Pivot Table Creation

Create a pivot table showing:

Rows → Region

Columns → Product

Values → Total Sales

Which region sells the most laptops?

```python
# Create pivot table
pivot_table = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)
print(pivot_table)

print("*"*100)

#  most laptops selling region
max_laptop_selling_region = pivot_table['Laptop'].idxmax()
max_laptop_selling_value = pivot_table['Laptop'].max()

print(f"Region with highest Laptop sales: {max_laptop_selling_region}")
print(f"Total Laptop Sales: {max_laptop_selling_value}")
```

Q8 Profitability Analysis

Calculate the average profit for each product category.

Which category is most profitable?

```python
#average profit for each product category
avg_profit_category = df.groupby('Category')['Profit'].mean().round(2)
print(avg_profit_category)

# most profitable category
print(f"\n Most profitable category: {avg_profit_category.idxmax()}")
```

Q9 Discount Impact Study

Analyze the relationship between discount and profit.

Tasks:

Calculate the average profit for each discount level

Identify whether higher discounts lead to lower profit

```python

# average profit for each discount level
avg_profit_discount = df.groupby('Discount')['Profit'].mean().round(2).reset_index()
avg_profit_discount['Discount'] = (avg_profit_discount['Discount'] * 100).astype(int).astype(str) + '%'
print(avg_profit_discount)


```

*Identifying  whether higher discounts lead to lower profit*  
 - We can clearly analyse that As discount:  high , profit : low
 - High discounts → sometimes negative profit (loss)

Strategic Suggestion :    
 - discounts might be necessary to drive sales volume .
 - The company should restrict 15% and 20% discounts
 - If required, reserving them only for extremely high-volume purchases or less profitable items




Q10 End-to-End Data Wrangling Task

Perform the following workflow:

Filter the dataset where Sales > 3000

Group data by Region and Product

Calculate total sales

Reshape the result into a pivot table

Explain the business insights obtained from this analysis.

```python
filtered_df = df[df["Sales"] > 3000]
grouped_sales = filtered_df.groupby(["Region", "Product"])["Sales"].sum().reset_index()
pivot_result = grouped_sales.pivot(index="Region", columns="Product", values="Sales").fillna(0)
print(pivot_result)
```

Analysis :  
 - East region shows strong sales in Laptop, Phone and Tablet.
 - West region is strong in Keyboard.
 - South region is strong in Mouse.
 - High-value sales are spread across different products and regions.
